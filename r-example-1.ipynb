# Example Using R 3.4.3

## Summary of Iris dataset

Show summary statistics for [Edgar Anderson's Iris Data](https://en.wikipedia.org/wiki/Iris_flower_data_set) dataset:

```r
summary(iris)
```

## Create Plots

### Plot Lengths

Plot Sepal vs Petal length:

```r
library(ggplot2)

qplot(Sepal.Length, Petal.Length, data = iris, color = Species, size = Petal.Length)
```

### Plot Widths

Plot Sepal vs Petal widths:

```r
library(ggplot2)

qplot(Sepal.Width, Petal.Width, data = iris, color = Species, size = Sepal.Width)
```

### Histogram

This is an example of embedded R to generate a plot.

```r
options(repr.plot.width=5, repr.plot.height=5)

set.seed(12)

samples <- rexp(100, rate = 0.2)
s <- seq(0, max(samples), 0.1)
d <- dexp(s, rate = 0.2)

hist(samples, prob = TRUE, main = "", col = adjustcolor("purple", alpha = 0.5), ylim = c(0, max(d)))
lines(density(samples), lty = 2)
curve(dexp(x, rate = 0.2), lwd = 2, add = TRUE)
rug(samples)
```

### Line Plot

Another example using the [ggplot2](https://ggplot2.tidyverse.org/) library:

```r
options(repr.plot.width=5, repr.plot.height=5)

library(ggplot2)
 
set.seed(4321)
 
n <- 10000
means <- cumsum(rnorm(n)) / (1 : n)
 
qplot(x = 1:n, y = means, 
      geom = "line",
      colour = I("purple"),
      xlab = "sample size", ylab = "cumulative means",
      main = "Law of Large Numbers") +
geom_hline(yintercept = 0)
```
